```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
plt.title('sports watch data',loc='left') # loc=left,right,centre
plt.xlabel('average pulse') # loc=left,right,centre
plt.ylabel('calorie burnage',loc='bottom') # loc=bottom,top,centre for ylabel
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
font1={'family':'serif','color':'blue','size':30}
font2={'family':'serif','color':'red','size':15}
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
plt.title('sports watch data',fontdict=font1)
plt.xlabel('average pulse',fontdict=font2)
plt.ylabel('calorie burnage',fontdict=font2)
plt.grid()
#plt.grid(linestyle='-.',linewidth=3,color='pink')
#plt.grid(axis='y') axis='x' axis='both'
plt.show()
```

Subplots
=

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('sports watch data')
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.title('sports watch data')
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
plt.suptitle('Data')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,5))
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,3,1)
plt.plot(x,y)
plt.title('sports watch data')
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,3,2)
plt.plot(x,y)
plt.title('sports watch data')
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,3,3)
plt.plot(x,y)
plt.title('sports watch data')
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,3,4)
plt.plot(x,y)
plt.title('sports watch data')
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,3,5)
plt.plot(x,y)
plt.title('sports watch data')
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,3,6)
plt.plot(x,y)
plt.title('sports watch data')
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
plt.suptitle('Data')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors=np.array([0,10,20,30,40,45,50,55,60,70,80,90,100])
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y,c=colors,cmap='PuBuGn',alpha=0.5,s=sizes) #alpha=opacity
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randint(10,50,size=20)
y=np.random.randint(10,50,size=20)
colors=np.random.randint(10,50,size=20)
sizes=10*np.random.randint(10,50,size=20)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='Spectral')
plt.colorbar()
plt.show()
```

Bar Plot
=

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.barh(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.bar(x,y,color='cyan',width=0.4) #if bar then width,if barh then height,by default value is 0.8
plt.title('sports watch data')
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.bar(x,y,color='cyan',width=0.4,label="ABC")
x=np.array(['P','Q','R','S'])
y=np.array([1,3,5,7])
plt.bar(x,y,color='red',width=0.4,label="PQR")
plt.legend(loc='upper left')#by default upper right, to see other options then shift tab on legend
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
y=np.array([35,25,25,15])
mylabel=['apple','banana','cherry','dates']
mycolor=['red','yellow','magenta','brown']
plt.pie(y,labels=mylabel,colors=mycolor)
plt.title("Pie Chart")
plt.legend(title='Four Fruits')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
y=np.array([35,25,25,15])
mylabel=['apple','banana','cherry','dates']
mycolor=['red','yellow','magenta','brown']
plt.pie(y,labels=mylabel,colors=mycolor,autopct='%1.1f%%')
plt.title("Pie Chart")
plt.legend(title='Four Fruits')
plt.tight_layout()
plt.show()
```
